```python
data = pd.read_csv('C:/Users/taeh0/Desktop/my future/TH Kim/KBO-project/final_data/final.csv')
data = data.drop([data.columns[0],data.columns[1],data.columns[2],'ER','R','SV','HLD','SO','IBB'],axis=1)
#data['IP']=data['IP'][:][:3]

for i in range(len(data)):
    data['IP'][i] = data['IP'][i][:3]

    data['ERA'][i] = round(data['ERA'][i])
data
```
